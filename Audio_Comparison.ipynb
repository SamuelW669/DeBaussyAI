```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display

def load_audio(file_path, sr=22050):
    """Load an audio file and return the time series and sample rate."""
    y, sr = librosa.load(file_path, sr=sr)
    return y, sr

def compute_spectrogram(y, sr, n_fft=2048, hop_length=512):
    """Compute the spectrogram of an audio signal."""
    S = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length))
    S_db = librosa.amplitude_to_db(S, ref=np.max)
    return S_db, sr, hop_length

def compare_spectrograms(spec1, spec2):
    """Compare two spectrograms by computing the mean squared error."""
    # Ensure the spectrograms have the same shape
    min_shape = min(spec1.shape, spec2.shape)
    spec1 = spec1[:min_shape[0], :min_shape[1]]
    spec2 = spec2[:min_shape[0], :min_shape[1]]

    mse = np.mean((spec1 - spec2) ** 2)
    return mse

def plot_spectrogram(S_db, sr, hop_length, title):
    """Plot a spectrogram."""
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()

# Load two audio files
file1 = '/content/drive/MyDrive/MIDI4STRINGS_but_better/Training Data/Human_0.mp3'
file2 = '/content/drive/MyDrive/MIDI4STRINGS_but_better/Training Data/Human_1.mp3'

y1, sr1 = load_audio(file1)
y2, sr2 = load_audio(file2)

# Ensure both audio files have the same sample rate
if sr1 != sr2:
    raise ValueError("Sample rates do not match!")

# Compute spectrograms
spec1, sr1, hop_length1 = compute_spectrogram(y1, sr1)
spec2, sr2, hop_length2 = compute_spectrogram(y2, sr2)

# Compare the spectrograms
mse = compare_spectrograms(spec1, spec2)
print(f"Mean Squared Error between spectrograms: {mse}")

# Optionally, plot the spectrograms
plot_spectrogram(spec1, sr1, hop_length1, "Spectrogram of Audio 1")
plot_spectrogram(spec2, sr2, hop_length2, "Spectrogram of Audio 2")
```
