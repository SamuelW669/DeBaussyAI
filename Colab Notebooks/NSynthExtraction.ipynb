```python
import json
import shutil
import os
```

```python
# Paths to dataset
dataset_dir = "/content/nsynth-train"  # Replace with the dataset directory
json_path = os.path.join(dataset_dir, "train.json")  # Adjust for "valid.json" or "test.json" if needed
audio_dir = os.path.join(dataset_dir, "audio")
output_dir = "/content/drive/MyDrive/DeBaussyAI/Training Data/NSynth Data Extractions"  # Replace with desired output directory for acoustic strings

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Load metadata
with open(json_path, "r") as f:
    metadata = json.load(f)

# Filter acoustic strings
filtered_files = []
for note_id, attributes in metadata.items():
    if attributes["instrument_source_str"] == "acoustic" and attributes["instrument_family_str"] == "string":
        filtered_files.append(attributes["note_str"])

# Copy filtered audio files to output directory
for file_name in filtered_files:
    src_path = os.path.join(audio_dir, file_name + ".wav")
    dst_path = os.path.join(output_dir, file_name + ".wav")
    if os.path.exists(src_path):
        shutil.copy(src_path, dst_path)
        print(f"Copied: {file_name}.wav")

print(f"Extracted {len(filtered_files)} acoustic string audio files to {output_dir}")
```
