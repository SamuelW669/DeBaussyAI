```python
# Conversion dictionaries
note_to_number = {
    "A": 1,
    "A#": 1.5,
    "B": 2,
    "B#": 2.5,
    "C": 3,
    "C#": 3.5,
    "D": 4,
    "D#": 4.5,
    "E": 5,
    "E#": 5.5,
    "F": 6,
    "F#": 6.5,
    "G": 7,
    "G#": 7.5
}

number_to_note = {
    1: "A",
    1.5: "A#",
    2: "B",
    3.5: "B#",
    3: "C",
    3.5: "C#",
    4: "D",
    4.5: "D#",
    5: "E",
    5.5: "E#",
    6: "F",
    6.5: "F#",
    7: "G",
    7.5: "G#"
}


def note_to_num(note_str):
  try:
    base_note = note_str[:-1]  # Extract the note (e.g., 'A', 'A#')
    octave = int(note_str[-1])  # Extract the octave (e.g., '1')
    base_number = note_to_number[base_note]
    final_number = base_number + 7*(octave - 1)
    return final_number
  except:
    return -1

def num_to_note(note_num):
  try:
    # print(1)
    octave = note_num//7 + 1
    # print(octave)
    base_number = note_num - 7*(octave - 1)
    # print(base_number)
    base_note = number_to_note[base_number]
    # print(base_note)
    final_note = base_note + str(int(octave))
    return final_note
  except:
    return -1

print(num_to_note(5.5))
print(note_to_num("B#8"))
```
