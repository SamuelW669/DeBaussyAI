```python
import os
import numpy as np
import pretty_midi
import librosa
import soundfile as sf
import subprocess
from pydub import AudioSegment

def midi_to_mp3(midi_file_path, soundfont_path, wav_output_path):
    # Convert MIDI to WAV using fluidsynth
    subprocess.run([
        'fluidsynth', '-ni', soundfont_path, midi_file_path, '-F', wav_output_path, '-r', '44100'
    ], check=True)

def audio_to_mel_spectrogram(audio_file, mel_spectrogram_file, sr=44100):
    """
    Convert audio to Mel-spectrogram.

    :param audio_file: Path to the audio file.
    :param mel_spectrogram_file: Output path for the Mel-spectrogram file.
    :param sr: Sampling rate.
    """
    # Load audio file
    audio_data, _ = librosa.load(audio_file, sr=sr)

    # Convert to Mel-spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sr, n_mels=128)

    # Convert amplitude to decibels
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Save Mel-spectrogram as a .npy file
    np.save(mel_spectrogram_file, mel_spectrogram_db)

def midi_to_mel_spectrogram(midi_file, soundfont_file, mel_spectrogram_file):
    """
    Convert MIDI file directly to Mel-spectrogram.

    :param midi_file: Path to the MIDI file.
    :param mel_spectrogram_file: Output path for the Mel-spectrogram file.
    """
    # Temporary audio file path
    audio_file = 'temp_audio.wav'

    # Convert MIDI to audio
    midi_to_mp3(midi_file, soundfont_file, audio_file)

    # Convert audio to Mel-spectrogram
    audio_to_mel_spectrogram(audio_file, mel_spectrogram_file)

    # Clean up temporary audio file
    os.remove(audio_file)

# # Example usage
# midi_file = 'your_midi_file.mid'  # Replace with your MIDI file path
# soundfont_file = 'your_soundfont.sf2'  # Replace with your SoundFont file path
# mel_spectrogram_file = 'mel_spectrogram.npy'  # Output file for Mel-spectrogram

# # Convert MIDI to Mel-spectrogram
# midi_to_mel_spectrogram(midi_file, soundfont_file, mel_spectrogram_file)

# print(f"Mel-spectrogram saved to: {mel_spectrogram_file}")
```
