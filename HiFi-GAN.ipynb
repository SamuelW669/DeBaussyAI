```python
import torch
import librosa
import numpy as np
from hifi_gan import Generator

# Load the pre-trained HiFi-GAN model
hifi_gan = Generator()
hifi_gan.load_state_dict(torch.load('path_to_hifi_gan_model.pt', map_location='cpu')['generator'])
hifi_gan.eval()

def mel_to_audio(mel_spectrogram):
    with torch.no_grad():
        mel_spectrogram = torch.tensor(mel_spectrogram).unsqueeze(0)  # Add batch dimension
        audio = hifi_gan(mel_spectrogram).squeeze().cpu().numpy()  # Generate audio
    return audio

def save_audio(audio, path, sr=22050):
    librosa.output.write_wav(path, audio, sr)
    print(f'Audio saved to {path}')

# Example usage:
mel_spectrogram = np.load('path_to_mel_spectrogram.npy')  # Load your mel spectrogram here
audio = mel_to_audio(mel_spectrogram)
save_audio(audio, 'output_audio.mp3')
```
